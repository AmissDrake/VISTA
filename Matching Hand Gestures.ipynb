```python
import cv2
import numpy as np
import mediapipe as mp
import tensorflow as tf

# Load Pre-Trained ASL Model
MODEL_PATH = r"C:\Users\scsai\Downloads\ResNet152V2-ASL.h5\ResNet152V2-ASL.h5"  # Update this path
model = tf.keras.models.load_model(MODEL_PATH)

# Initialize Mediapipe for Hand Tracking
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=False, max_num_hands=1, min_detection_confidence=0.7)
mp_draw = mp.solutions.drawing_utils

# Start Webcam
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    h, w, _ = frame.shape  # Get frame dimensions
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(rgb_frame)

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            # Get bounding box around hand
            x_min = int(min(lm.x for lm in hand_landmarks.landmark) * w)
            y_min = int(min(lm.y for lm in hand_landmarks.landmark) * h)
            x_max = int(max(lm.x for lm in hand_landmarks.landmark) * w)
            y_max = int(max(lm.y for lm in hand_landmarks.landmark) * h)

            # Ensure valid coordinates
            x_min, y_min = max(0, x_min), max(0, y_min)
            x_max, y_max = min(w, x_max), min(h, y_max)

            # Crop & resize hand region
            hand_region = frame[y_min:y_max, x_min:x_max]
            hand_region = cv2.resize(hand_region, (256, 256))
            hand_region = np.expand_dims(hand_region, axis=0)  # Add batch dimension
            hand_region = hand_region / 255.0  # Normalize

            # Predict ASL letter
            prediction = model.predict(hand_region)
            detected_letter = chr(np.argmax(prediction) + 65)  # Convert to ASCII A-Z

            # Display predicted letter
            cv2.putText(frame, f"Letter: {detected_letter}", (50, 100),
                        cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 3)

            # Draw hand landmarks
            mp_draw.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

    cv2.imshow("ASL Recognition", frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
